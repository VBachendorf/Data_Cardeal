## Exemplos de Código

### Laço `for`

```python
# Imprimindo números de 1 a 5
for i in range(1, 6):
    print(i)

# Percorrendo uma lista
frutas = ["maçã", "banana", "laranja"]
for fruta in frutas:
    print(fruta)
